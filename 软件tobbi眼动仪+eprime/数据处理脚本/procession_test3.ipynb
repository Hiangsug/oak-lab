# RawData处理

```python
import pandas as pd
import numpy as np
```

```python
filename = 'leguan_test3_Rec 32.tsv'
data1 = pd.read_csv(filename,sep='\t')  
```

```python
data2 = data1.drop(['GazeEventType','RecordingName','RecordingDate','FixationFilter','RecordingTimestamp'],axis='columns')
index1 = data2['MediaName'].dropna().index
print(index1)#非空名的index
```

```python
index1 = data2['MediaName'].dropna().index
print(index1)#非空名的index
```

```python
data3 = data2[data2.index.isin(index1)]
data3.head()
```

```python
data3.tail()
```
