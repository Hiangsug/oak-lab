```python
import pandas as pd
filename = 'leguan_test1_Rec 36.tsv'
data1 = pd.read_csv(filename,sep='\t')  
data2 = data1.drop(['GazeEventType','RecordingName','RecordingDate','FixationFilter','RecordingTimestamp'],axis='columns')
index1 = data2['MediaName'].dropna().index
index1 = data2['MediaName'].dropna().index
data3 = data2[data2.index.isin(index1)]
data3.head()
index2 = data3[data3['MediaName']=='fixation.jpg'].index
data4 = data3[~data3.index.isin(index2)]
data4.head() #删除多余图片刺激
```

```python
group0101 = data4[data4['MediaName']=='01-10.jpg']  #0100组合数据
group0101 = group0101.dropna(how='all',axis = 'columns')  #删除竖排缺省值
group0101.head()
```

```python
group0101 = group0101.rename(columns={'AOI[Rectangle]Hit.2': 'left2','AOI[Rectangle 2]Hit.2': 'right2'})
group0101 = group0101.rename(columns={'AOI[Rectangle]Hit.7': 'left3','AOI[Rectangle 2]Hit.7': 'right3'})
group0101 = group0101.rename(columns={'AOI[Rectangle]Hit.10': 'left4','AOI[Rectangle 2]Hit.10': 'right4'})
group0101 = group0101.rename(columns={'AOI[Rectangle]Hit.20': 'left5','AOI[Rectangle 2]Hit.20': 'right5'})
group0101.head()    #修改列标
```

```python
group0101 = group0101.drop_duplicates()
group0101.head()
```
